### Pylot API vs Object-Oriented(OO) API
* Pyplot API -> quickly
* OO API -> advanced

Pylot API

```python
x=np.linspace(0,10,100)
plt.plot(x,np.sin(x),color="blue",linestyle="dashed",label="sin(x)")
plt.plot(x,np.cos(x),color="red",label="cos(x)")
plt.title("Sin(x) and Cos(X)")
plt.xlabel("X")
plt.ylabel("F(X)")
# plt.xlim([0,4]) giới hạn
# plt.ylim([-0.7,0.7]) giới hạn
# plt.axis([0,4,-0.7,0.6]) giống xlim và ylim
# plt.axis("tight")
plt.axis("equal")
plt.legend();# làm cho nó fit chắc với đồ thị
```

Object-Oriented API

```python
# 1.prepare data
x=[1,2,3,4]
y=[11,22,33,44]
# 2. Setup plot
fig, ax=plt.subplots(figsize=(5,5))#Figure size = width&height od the plot
# 3. Plot data
ax.plot(x,y)

ax.set(title="A simple plot", xlabel="x-axis",ylabel="y-axis")
```

## 2. Most common types Matplotlib plots

Line

```python
# create an array
x=np.linspace(0,10,100)
x[:5]
```

```python
# 00 API plot a line chart
fig, ax=plt.subplots()
ax.plot(x,x**3)
```

Scatter

```python
# Pylot API
plt.scatter(x,np.exp(x))
```

```python
# 00 API
# Prepare data
rng=np.random.RandomState(0)
x=rng.randn(100)
y=rng.randn(100)
colors=rng.rand(100)
sizes=1000*rng.rand(100)
# Plot
fig,ax=plt.subplots()
img1=ax.scatter(x,y,s=sizes,c=colors,cmap="viridis",alpha=0.3)

fig.colorbar(img1)
```

### Bar
* vertical
* Horizontal

```python
# Prepare data from a python dictionary
soft_drink_prices={"coca":10,
                   "pepsi":12,
                   "sprite":8}
fig,ax=plt.subplots()
ax.bar(soft_drink_prices.keys(),soft_drink_prices.values())
ax.set(title="Bách hoá xanh soft drink privce",ylabel="price ($)")
```

```python
fig, ax=plt.subplots()
ax.barh(list(soft_drink_prices.keys()),list(soft_drink_prices.values()))
```

### Histogram

```python
# prepare data
np.random.seed(42)
studen_height=np.random.normal(170,10,250)
```

```python
#pylot API
plt.hist(studen_height)
```

### Subplot

```python
# Option 1:
fig, ((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2,figsize=(10,5))
# Plot data to each axis
ax1.plot(x,x/2)
ax1.plot(x,x*2)
ax2.scatter(np.random.random(10),np.random.random(10))
ax3.bar(soft_drink_prices.keys(),soft_drink_prices.values())
ax4.hist(studen_height)
#
```

```python
# Option 2:
fig, ax=plt.subplots(nrows=2,ncols=2,figsize=(10,5))
# Plot data to each axis
ax[0,0].plot(x,x/2)
ax[0,0].plot(x,x*2)
ax[0,1].scatter(np.random.random(10),np.random.random(10))
ax[1,0].bar(soft_drink_prices.keys(),soft_drink_prices.values())
ax[1,1].hist(studen_height)
```
